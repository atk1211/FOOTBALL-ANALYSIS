```python
!pip install mplsoccer
!pip install highlight_text
import pandas as pd
import matplotlib.pyplot as plt
from highlight_text import fig_text
import matplotlib as mpl
from mplsoccer.pitch import Pitch, VerticalPitch
import seaborn as sns

passes = pd.read_csv("/content/Passes.csv", delimiter=";")
shots = pd.read_csv("/content/Shots.csv", delimiter=";")
```

  **PASSES ANALYSIS**

```python
passes['x'] = passes['x']*1.2
passes['y'] = passes['y']*.8
passes['endX'] = passes['endX']*1.2
passes['endY'] = passes['endY']*.8

passes
```

```python
fig ,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')
pitch = Pitch(pitch_type='statsbomb',
              pitch_color='grass', line_color='white', stripe=True)
pitch.draw(ax=ax)
plt.gca().invert_yaxis()
kde = sns.kdeplot(
        data=passes[['x', 'y']],
        shade = True,
        shade_lowest=False,
        alpha=.5,
        n_levels=10,
)
for x in range(len(passes['x'])):
    if passes['outcome'][x] == 'Successful':
        plt.plot((passes['x'][x],passes['endX'][x]),(passes['y'][x],passes['endY'][x]),color='green')
        plt.scatter(passes['x'][x],passes['y'][x],color='green')
    if passes['outcome'][x] == 'Unsuccessful':
        plt.plot((passes['x'][x],passes['endX'][x]),(passes['y'][x],passes['endY'][x]),color='red')
        plt.scatter(passes['x'][x],passes['y'][x],color='red')
plt.xlim(0,120)
plt.ylim(0,80)
plt.title('Messi Pass Map vs Real Betis',color='white',size=20)
fig.tight_layout()

plt.show()
```

```python

fig, ax = plt.subplots(figsize=(13,8.5))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')
pitch = VerticalPitch(pitch_type='statsbomb',
                      pitch_color='grass', line_color='white',
                      stripe=True, half=True)
pitch.draw(ax=ax)
plt.gca().invert_yaxis()
plt.scatter(shots.x, shots.y, c='red', s=100, alpha=.8)
plt.title('Barcelona Shot Chart vs Juventus', fontsize=24, color='w')
total_shots = len(shots)
fig_text(s=f'Total Shots: {total_shots}',
        x=.27, y =.67, fontsize=14, fontfamily='DejaVu Sans', color='w')
fig_text(s=f'xG: .85',
        x=.49, y =.67, fontsize=14, fontfamily='DejaVu Sans', color='w')
fig_text(s=f'Goals: 0',
        x=.68, y =.67, fontsize=14, fontfamily='DejaVu Sans', color='w')

plt.show()
```
